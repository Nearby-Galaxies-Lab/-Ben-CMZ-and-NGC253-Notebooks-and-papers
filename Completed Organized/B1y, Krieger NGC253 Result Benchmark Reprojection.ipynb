# 6.1 Splicing

```python
FOV = [70,360]#pc
gal="NGC253"
Min_res=4.3*u.pc#Smallest resolution


iterations = 6
Line_Name = '_NGC253_CO_J3_2_'
Side=""
ovs = 3 #over sample factor for beam
min_vel = 3.3#km/s
iter_factor = 1/5

Initial_vel=1
    
    
for kl in range(5,6):
    Prime_Beam = Min_res*((kl*iter_factor+1))#Beam_Sizes[kl]
    Smoothe_4 = False
    Match_to_HCO = False

    Smoothe_2_5 =False
    
    if(kl==5):
        Prime_Beam = 3*u.pc#Nico
        ovs=5
        FOVp=[400,800]
        Smoothe_2_5=True
        Force_Origins=False
        
    else:
        ovs=3
        FOVp=FOV
    if(Smoothe_2_5):
        Smoothe_title="Smoothe_to_2.5"
    elif(Smoothe_4):
        Smoothe_title="G_width_4"
    elif(Match_to_HCO):
        Smoothe_title='Smoothe_to_3.3'
    else:
        Smoothe_title=''
    
    #Reproject using the function

    
    Cube_Name_Save = str(Prime_Beam.value)+"pc_beam_"+Line_Name+str(FOVp[0])+"x"+str(FOVp[1])+'pc_'+Side+'_vel_res_NEW'+Smoothe_title+'.fits'

    #Splice_vels(Cube_Name_Save)
    print('spliced')
    
    for km in range(5,6):
        vel_prime = min_vel*((km*iter_factor+1))
        if(km==5):
            vel_prime=3.3
        if(Smoothe_2_5):
            vel_prime=2.5
        Cube_Name_Load=Cube_Name_Save
        
        Cube_Name_Save = str(Prime_Beam.value)+"pc_beam_"+Line_Name+str(FOVp[0])+"x"+str(FOVp[1])+'pc_'+str(vel_prime)+Side+'_vel_res_NEW'+Smoothe_title+'.fits'
        
        sc = SpectralCube.read(Cube_Name_Load) #[:] 
        
        if(gal=="GC"):
            vel = np.arange(-325,326,vel_prime)*u.km/u.s    
        elif gal=="NGC253":
            vel = np.arange(-0,501,vel_prime)*u.km/u.s    


        if(Smoothe_2_5):
            G_width = np.sqrt(2.5**2-Initial_vel**2)
        if(Match_to_HCO):
            G_width = np.sqrt(3.3**2-Initial_vel**2)
        if(Smoothe_4):
            G_width = 4

        #G_width = 2.5**2
        G_width = 2.5

        Repo_Velocity(scp=sc,Cube_Name_Save=Cube_Name_Save,spec_grid=vel,G_width=G_width)
        del sc
        gc.collect()



    gc.collect()######################################################################
    print("done")
```
